```python
designation=[]
company_name=[]
rating=[]
location=[]
description=[]
day_posted=[]
salary=[]
for i in range(1,410,10):
    URL = "https://in.indeed.com/jobs?q=data%20science&l&vjk=745a3a4c86f3f023".format(i)
    page=requests.get(URL)
    htmlCode=page.text
    soup=BeautifulSoup(htmlCode)
    for x in soup.find_all('div',attrs={'class':'job_seen_beacon'}):
        post= x.find('h2',class_='jobTitle')
        if post is None:
            designation.append(np.NaN)
        else:
            designation.append(post.text)
        cm= x.find('span',class_='companyName')
        if cm is None:
            company_name.append(np.NaN)
        else:
            company_name.append(cm.text)
        rate=x.find('span',class_='ratingNumber')
        if rate is None:
            rating.append(np.NaN)
        else:
            rating.append(rate.text)
        loc=x.find('span',class_='companyLocation')
        if loc is None:
            location.append(np.NaN)
        else:
            location.append(loc.text)
        desc=x.find('div',class_='job-snippet')
        if desc is None:
            description.append(np.NaN)
        else:
            description.append(desc.text)
        post_date=x.find('span',class_='date')
        if post_date is None:
            day_posted.append(np.NaN)
        else:
            day_posted.append(post_date.text)
        sal=x.find('div',class_='metadata salary-snippet-container')
        if sal is None:
            salary.append(np.NaN)
        else:
            salary.append(sal.text)
        
```
